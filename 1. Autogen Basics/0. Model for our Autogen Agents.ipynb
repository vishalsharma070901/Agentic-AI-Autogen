```python
import sys
print(sys.executable)

import autogen_core
print(autogen_core.__file__)
print(autogen_core.__version__)
```

```python
from autogen_core.models import UserMessage
from autogen_ext.models.openai import OpenAIChatCompletionClient
import os
import dotenv
dotenv.load_dotenv()

groq_model_client = OpenAIChatCompletionClient(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1",
    model_info={
        "family": "llama",
        "vision": False,
        "function_calling": True,
        "json_output": True,
        "structured_output": True,
    },
)

response = await groq_model_client.create(
    [
        UserMessage(
            content="explain about reactjs in 100 words",
            source="user"
        )
    ]
)

print(response.content)

await groq_model_client.close()
```
